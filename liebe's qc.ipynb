```python
#this code will create a quantum circuit of 5 qubits
import matplotlib.pyplot as plt     #Imports the Matplotlib plotting library.
                                    #pyplot is the part of Matplotlib used to create graphs.
                                    #plt is just a short alias.
from qiskit import QuantumCircuit   #Imports the QuantumCircuit class so you can create quantum circuits.
from qiskit_aer import AerSimulator #Imports the AerSimulator, which behaves like a virtual quantum computer.
from qiskit.visualization import plot_histogram #Imports a function that draws a bar chart of measurement results
qc = QuantumCircuit(5)  #creates a quantum circuit of 5 qubits
qc.h(0)   #this add hadamard gate to the first qubit q0, which creates a superposition between 0 and 1
qc.x(1)   #this add not gate to second qubit q1, which flips 0 to 1 and vice versa
qc.y(2)   #this add pauli y gate to third qubit q2, which flips the qubit and add a phase change of 180
qc.z(3)   #this add pauli z gate to fourth qubit q3, which only changes the phase of qubit by 180 
qc.s(4)   #this gate changes the phase of the fifth qubit by 90
qc.t(0)   #this gate changes the phase of the first qubit by 45
qc.h(2)   #this add h gate to third qubit
qc.x(3)   #this add x gate to fourth
qc.y(4)   #this add y gate to fifth
qc.z(1)   #this add z gate to the second
qc.cx(0, 1)   #this create a quantum entanglement between first and second qubit
qc.cx(2, 3)   #this create entanglement between third and fourth 
qc.measure_all()  #this measure the every qubit
print("Quantum Circuit:")  #print the circuit
qc.draw("mpl")  #this create a diagram of the circuit
simulator = AerSimulator() #this create a virtual quantum computer, a simulator
result = simulator.run(qc, shots=1000).result() #the simulator measures the qubits 1000 times
counts = result.get_counts() #generate the counts
print("Measurement Counts:") #shows the result
print(counts)
plot_histogram(counts) #creates a histogram
```
